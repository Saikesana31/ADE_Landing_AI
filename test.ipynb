```python
# Load the api key 
import os
from dotenv import load_dotenv
from landingai_ade import LandingAIADE

load_dotenv()

vision_agent_api_key = os.getenv("VISION_AGENT_API_KEY")

# Initialize the client
client = LandingAIADE()
```

```python
# Parse - converts documents into structured markdown with chunk and grounding metadata
from pathlib import Path

# Load the documents
response = client.parse(
    document = Path("data/Mastering_Multi_Agent_Systems__1762311738.pdf"),
    model = "dpt-2-latest")

print(response.chunks)

# save the markdown file 
with open("texts.md", "w") as f:
    f.write(response.markdown)
    
```
